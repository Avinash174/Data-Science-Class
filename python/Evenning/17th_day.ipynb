```python
import pandas as pd 
```

```python
data1={
    "Emp_id":["E01","E02","E03","E04","E05"],
    "Name":["Ram","Sham","Avinash","Rahul","Ravi"],
    "Age":[34,90,10,75,10]
}

data2={
    "Emp_id":["E01","E02","E03","E04","E05"],
    "Salary":[4500,2344,1880,7499,8460]
}

df4=pd.DataFrame(data1)
df5=pd.DataFrame(data2)
```

```python
print(df4)
print()
print(df5)
```

```python
pd.merge(df4,df5,on="Emp_id") # join the basis of emp id
```

```python
pd.merge(left=df4,right=df5,on="Emp_id",how="left")
#left join 
```
